```python
import geopandas as gpd
```

```python
path = "data/france/data_10.gpkg"

# Load the data
dax = gpd.read_file(path)
data = dax.to_crs(epsg=4326)

data.head()
```

```python
# Get the max and min values for the coordinates
max_x = data.bounds.maxx.max()
min_x = data.bounds.minx.min()
max_y = data.bounds.maxy.max()
min_y = data.bounds.miny.min()
```

```python
max_x, min_x, max_y, min_y
```

```python
x = [[max_y, min_x], [max_y, max_x], [min_y, max_x], [min_y, min_x]]

y = []
for i in x:
    y.append([i[1],i[0]])
    
y
```
